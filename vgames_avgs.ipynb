```python
import pandas as pd
import numpy as np

#video game data source 1
vg = pd.read_csv('vgsales.csv')
def get_shape_and_column(vg):
    data = pd.DataFrame(data=vg)
    return data
```

```python
get_shape_and_column(vg)
```

```python
vg["Game Title"] = vg["Name"]
get_shape_and_column(vg)
```

```python
#video game data source 2
vg2 = pd.read_csv('Video Games Sales.csv')
def get_shape_and_column(vg2):
    data = pd.DataFrame(data=vg2)
    return data
```

```python
get_shape_and_column(vg2)
```

```python
#show what columns vg and vg2 have in common
vg["Rank1"] = vg["Rank"]
vg2["Rank2"] = vg2["Rank"]
column_names=[]
for column in vg:
    if column in vg2:
        column_names.append(column)
print(column_names)
```

```python
#merge vg and vg2 based on columns Game Title, Year and Platform to ensure data is merged with the proper row(some games have the same name released different years with different sales data)
import numpy as np
df = pd.merge(vg, vg2, on=["Game Title","Year","Platform"],suffixes=('', '_remove'))
df.drop([i for i in df.columns if 'remove' in i],
               axis=1, inplace=True)
#drop duplicate columns that represent the same data under different column names(example Name and Game Title)
df.drop('Name', axis=1, inplace=True)
df.drop('index', axis=1, inplace=True)
#Rank to be an average of both ranks
cols = ['Rank1','Rank2']
df['Rank'] = df[cols].astype(float).mean(axis=1)
df['Rank'] = df['Rank'].astype(int)
```

```python
#checking column data types to help with formatting(example get rid of decimals in year by changing to int)
df['Year'] = df['Year'].astype("Int64")
print(df.dtypes)
```

```python
#just checking
df.head()
```

```python
#Averaging Sales
cols = ['NA_Sales','North America']
df['NA_avg'] = df[cols].astype(float).mean(axis=1)


cols = ['EU_Sales','Europe']
df['EU_avg'] = df[cols].astype(float).mean(axis=1)


cols = ['JP_Sales','Japan']
df['JP_avg'] = df[cols].astype(float).mean(axis=1)


cols = ['Other_Sales','Rest of World']
df['Other_avg'] = df[cols].astype(float).mean(axis=1)


cols = ['Global_Sales','Global']
df['Global_avg'] = df[cols].astype(float).mean(axis=1)
```

```python
df.sort_values(by=['Rank'])
```

```python
#droping columns used to average
df.drop('NA_Sales', axis=1, inplace=True)
df.drop('EU_Sales', axis=1, inplace=True)
df.drop('JP_Sales', axis=1, inplace=True)
df.drop('Other_Sales', axis=1, inplace=True)
df.drop('Global_Sales', axis=1, inplace=True)
df.drop('Rank1', axis=1, inplace=True)
df.drop('Rank2', axis=1, inplace=True)
df.drop('North America', axis=1, inplace=True)
df.drop('Europe', axis=1, inplace=True)
df.drop('Japan', axis=1, inplace=True)
df.drop('Rest of World', axis=1, inplace=True)
df.drop('Global', axis=1, inplace=True)

#and redoing rank so it matches the index number
df['Rank_avg'] = df.index + 1
df.drop('Rank', axis=1, inplace=True)
display(df)

```

```python
#last but not least reorder columns
df1=df[df.columns[[11,5,1,2,3,4,6,7,8,9,10]]]
```
